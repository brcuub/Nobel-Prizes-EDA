```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 7]

ax = sns.lineplot(x="decade", y="award_mean", data=data2)

from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter())
```

- 1901 yılından bu yana gerçekleştirilen Nobel Ödül Töreninde en çok ödül alan ülke olarak öne çıkan ülke "United States of America" olmuştur. Veriler dikkatle incelendiğinde United States of America'nın 1930'lu yıllara kadar Nobel Ödüllerindeki hakimiyeti sınırlı iken bu yıllardan itibaren hakimiyetini arttırdığı görülmektedir. ABD'nin 1930'lu yıllardan itibiren hakimiyetini arttırması II. Dünya Savaşı'na katılması ve askeri ve tıbbi ihtiyaçlarını karşılamak için bu alanda gelişmeler göstermesi ile açıklanabilir.

```python
awards_categorically = data1[data1['birth_country'] == most_awarded_country].groupby(['category', 'year']).size().reset_index(name='award_count')
```

```python
# ABD'nin Kategori Bazında Yıllar İçinde Aldığı Ödüllerin Zaman Serisi Grafiği
categorical_charts= awards_categorically.groupby(["category", "year"]).agg({"award_count": "mean"}).reset_index()

categories = categorical_charts['category'].unique()
num_plots = len(categories)
num_cols = 2
num_rows = -(-num_plots // num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 15))

# Her bir kategori için çizgi grafiği oluşturun
for i, category in enumerate(categories):
    category_data = categorical_charts[categorical_charts['category'] == category]
    ax = axes[i // num_cols, i % num_cols]
    ax.plot(category_data['year'], category_data['award_count'], marker='o', color='red')
    ax.set_title(f"{category} Kategorisi")
    ax.set_xlabel('Yıl')
    ax.set_ylabel('Ödül Ortalaması')
    ax.grid(True)

# Eksik subplot'ları gizleyin
for i in range(num_plots, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

ABD'nin en çok ödül alan ülke olarak öne çıkmasında etkili olan faktörleri anlamak için her bir kategoride zaman içinde elde ettiği başarıları incelemek doğru olacaktır.

```python
# Kategorilere Göre Cinsiyet Dağılımı
data1.groupby("category")["sex"].value_counts() * 100 / data1.groupby("category")["sex"].size()
```

```python
data1['decade'] = (data1['year'] // 10) * 10
decade_by_sex = data1.groupby(["category",'decade'])['sex'].value_counts(normalize=True) * 100
decade_by_sex
```

```python
decade_by_sex_df = decade_by_sex.reset_index()
decade_by_sex_df[decade_by_sex_df["category"]=="Peace"] 
```

```python
categories = data1['category'].unique()
num_plots = len(categories)
num_cols = 2
num_rows = -(-num_plots // num_cols)

# Alt grafiklerin oluşturulması
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 15))

for i, category in enumerate(categories):
    # Her bir kategori için cinsiyet dağılımını 10'ar yıllık dönemler halinde hesaplayın
    category_data = data1[data1['category'] == category]
    decade_by_sex = category_data.groupby(['decade', 'sex']).size().unstack(fill_value=0)
    decade_by_sex = decade_by_sex.apply(lambda x: x / x.sum(), axis=1) * 100
    
    # Grafiği çizdirin
    ax = axes[i // num_cols, i % num_cols]
    decade_by_sex.plot(kind='line', ax=ax, marker='o')
    ax.set_title(f'{category} Kategorisi Cinsiyet Dağılımı (10 Yıllık Dönemler)')
    ax.set_xlabel('Decade')
    ax.set_ylabel('Percentage (%)')
    ax.yaxis.set_major_formatter(PercentFormatter())
    ax.grid(True)

plt.tight_layout()
plt.show()
```

- Yıllar içinde Nobel  Ödüllerinde  kadın erkek dağılımı incelendiğinde  kadınların bir çok kategoride başarılarının ilk günden bu yana hep sınırlı kaldığı dikkat çekmektedir.
